```python
def copy_images(source_dir, dest_dir, num_images):
    """Copies a specified number of images from the source directory to the destination directory."""
    image_extensions = ('.jpg', '.jpeg', '.png', '.gif', '.bmp')
    copied_count = 0
    for filename in os.listdir(source_dir):
        if copied_count >= num_images:
            break
        if filename.lower().endswith(image_extensions):
            source_path = os.path.join(source_dir, filename)
            dest_path = os.path.join(dest_dir, filename)
            try:
                shutil.copy2(source_path, dest_path) # copy2 preserves metadata
                copied_count += 1
                print(f"Copied: {filename}")
            except shutil.SameFileError:
                print(f"Skipping: {filename} (same file)")
            except Exception as e:
                print(f"Error copying {filename}: {e}")
    print(f"Copied {copied_count} images from {source_dir} to {dest_dir}")


# Example usage: Copy 100 images from dataset_path_1 to augmented_output_path
copy_images(dataset_path_1, augmented_output_path, 100)

# Example usage: Copy 50 images from dataset_path_2 to augmented_output_path
copy_images(dataset_path_2, augmented_output_path, 50)
```

```python
# Function to count images per class in each dataset split (train/test/valid)
def count_images_per_class(dataset_path):
    class_counts = {}

    for split in ["train", "test", "valid"]:
        split_path = os.path.join(dataset_path, split)

        if os.path.exists(split_path):
            for class_folder in os.listdir(split_path):
                class_path = os.path.join(split_path, class_folder)

                if os.path.isdir(class_path):
                    num_images = len(os.listdir(class_path))
                    class_counts[class_folder] = class_counts.get(class_folder, 0) + num_images

    return class_counts
```

```python
# Function to copy images from source dataset to the output folder
def merge_augmented_images(source_path, destination_path):
    for split in ["train", "test", "valid"]:  # Iterate over train, test, valid splits
        split_source = os.path.join(source_path, split)
        split_dest = os.path.join(destination_path, split)

        if os.path.exists(split_source):
            os.makedirs(split_dest, exist_ok=True)

            for class_folder in os.listdir(split_source):  # Iterate over class folders
                class_source = os.path.join(split_source, class_folder)
                class_dest = os.path.join(split_dest, class_folder)

                if os.path.isdir(class_source):
                    os.makedirs(class_dest, exist_ok=True)

                    # Copy images
                    for img in os.listdir(class_source):
                        src_img_path = os.path.join(class_source, img)
                        dst_img_path = os.path.join(class_dest, img)

                        # Avoid duplicate filenames
                        if os.path.exists(dst_img_path):
                            base, ext = os.path.splitext(img)
                            new_filename = f"{base}_copy{ext}"
                            dst_img_path = os.path.join(class_dest, new_filename)

                        shutil.copy(src_img_path, dst_img_path)
```

```python
import os
import random
import shutil

def split_data(source_dir, train_ratio=0.7, val_ratio=0.15, test_ratio=0.15):
    """Splits data into train, validation, and test sets.

    Args:
        source_dir: Path to the source directory containing class folders.
        train_ratio: Proportion of data for training.
        val_ratio: Proportion of data for validation.
        test_ratio: Proportion of data for testing.
    """

    if train_ratio + val_ratio + test_ratio != 1.0:
        raise ValueError("Ratios must sum to 1.0")

    for class_folder in os.listdir(source_dir):
        class_path = os.path.join(source_dir, class_folder)

        if os.path.isdir(class_path):
            images = [f for f in os.listdir(class_path) if os.path.isfile(os.path.join(class_path, f))]
            random.shuffle(images)

            num_images = len(images)
            num_train = int(num_images * train_ratio)
            num_val = int(num_images * val_ratio)

            train_images = images[:num_train]
            val_images = images[num_train:num_train + num_val]
            test_images = images[num_train + num_val:]


            for split, image_subset in zip(["train", "valid", "test"], [train_images, val_images, test_images]):
              split_path = os.path.join(augmented_output_path, split, class_folder)
              os.makedirs(split_path, exist_ok = True)
              for image in image_subset:
                  source_image_path = os.path.join(class_path, image)
                  destination_image_path = os.path.join(split_path, image)
                  shutil.copy(source_image_path, destination_image_path)


# Example usage (replace with your source directory and desired ratios):
source_directory = "/content/drive/MyDrive/cse438PumkinDS/Augmented"
split_data(source_directory)
```

```python
import os

augmented_output_path = "/content/drive/MyDrive/cse438PumkinDS/Augmented"

def check_split(dataset_path):
    """Checks if the dataset is correctly split into train, test, and validation sets."""
    for split in ["train", "test", "valid"]:
        split_path = os.path.join(dataset_path, split)
        if not os.path.exists(split_path):
            print(f"âŒ Error: '{split}' split directory not found at {split_path}")
            return False

        #Check if each split contains subfolders
        if not os.listdir(split_path):
            print(f"âŒ Error: '{split}' split directory is empty at {split_path}")
            return False

        for class_folder in os.listdir(split_path):
            class_path = os.path.join(split_path, class_folder)
            if not os.path.isdir(class_path):
                print(f"âŒ Error: '{class_folder}' is not a directory in {split} split")
                return False

            if not os.listdir(class_path):
                print(f"âŒ Error: '{class_folder}' in '{split}' split is empty")
                return False
    print("âœ… Dataset is correctly split into train, test, and validation sets.")
    return True

check_split(augmented_output_path)
```

ResneXT-50

```python
import torch
import torchvision.models as models
from torch import nn

# Load pre-trained ResNeXt-50 (weights from ImageNet)
model = models.resnext50_32x4d(pretrained=True)


num_classes = 5  # Adjust based on pumpkin leaf classes
model.fc = nn.Linear(model.fc.in_features, num_classes)

device = torch.device("cpu")
model = model.to(device)
```

```python
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)  # Lower learning rate for fine-tuning
```

```python
from torch.utils.data import DataLoader

# Define batch size (small for CPU)
batch_size = 5

# Create data loaders
train_loader = DataLoader(
    train_dataset,
    batch_size=batch_size,
    shuffle=True,
    num_workers=0   # Set to 0 for CPU (no multiprocessing)
)

val_loader = DataLoader(
    val_dataset,
    batch_size=batch_size,
    shuffle=False,  # No need to shuffle validation
    num_workers=0
)
```

```python
epochs = 3  # Start with a small number
best_val_accuracy = 0.0

for epoch in range(epochs):
    model.train()
    running_loss = 0.0
    progress_bar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}")

    for images, labels in progress_bar:
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        progress_bar.set_postfix(loss=running_loss / len(train_loader))

    # Validation
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in val_loader:
            outputs = model(images)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    val_accuracy = 100 * correct / total
    print(f"Val Accuracy: {val_accuracy:.2f}%")

    # Save best model
    if val_accuracy > best_val_accuracy:
        best_val_accuracy = val_accuracy
        torch.save(model.state_dict(), "best_resnext50_pumpkin.pth")
```

```python
#calculate recall precision f1 score for this model

from sklearn.metrics import recall_score, precision_score, f1_score

y_true = [] # True labels
y_pred = [] # Predicted labels

# Accumulate true and predicted labels during model evaluation
model.eval()
with torch.no_grad():
    for images, labels in val_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)
        y_true.extend(labels.cpu().numpy())
        y_pred.extend(predicted.cpu().numpy()) #

# Calculate recall, precision, and F1-score
recall = recall_score(y_true, y_pred, average='weighted')  # Use 'macro', 'micro' or other averages as needed
precision = precision_score(y_true, y_pred, average='weighted')
f1 = f1_score(y_true, y_pred, average='weighted')

print(f"Recall: {recall}")
print(f"Precision: {precision}")
print(f"F1 Score: {f1}")
```

DenseNet-201

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import DenseNet201
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
import os
```

```python
# Define dataset paths
dataset_path_1 = "/content/drive/MyDrive/Pumpkin Leaf Diseases Dataset From Bangladesh/Augmented"
dataset_path_2 = "/content/drive/MyDrive/sweet pumpkin diseas/Augmented Images/Augmented Sweet Pumpkin Downy Mildew Disease"
augmented_output_path = "/content/drive/MyDrive/cse438PumkinDS/Augmented"  # Output path

# Combine dataset paths if necessary
train_data_dir = augmented_output_path  # Use the preprocessed dataset
```

```python
# Image parameters
img_size = (224, 224)  # Input size for DenseNet201
batch_size = 5

# Define ImageDataGenerator for training and validation
train_datagen = ImageDataGenerator(
    rescale=1.0/255,
    validation_split=0.2  # 80-20 train-validation split
)

train_generator = train_datagen.flow_from_directory(
    train_data_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',  # Use 'categorical' for multi-class classification
    subset='training'
)

val_generator = train_datagen.flow_from_directory(
    train_data_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation'
)
```

```python
# Load DenseNet201 pre-trained model (without top layers)
base_model = DenseNet201(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Freeze the base model layers
base_model.trainable = False  # Set to True if fine-tuning later
```

```python
# Add custom layers on top of DenseNet201
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.5)(x)  # Dropout to prevent overfitting
x = Dense(256, activation='relu')(x)
x = Dropout(0.3)(x)

# Output layer
num_classes = len(train_generator.class_indices)  # Auto-detect number of classes
output_layer = Dense(num_classes, activation='softmax')(x)

# Create the final model
model = Model(inputs=base_model.input, outputs=output_layer)
```

```python
model.compile(
    optimizer=Adam(learning_rate=0.0001),  # Lower LR for better convergence
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# Train the model
epochs = 3
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=epochs,
    verbose=1
)
# Evaluate on validation set
val_loss, val_acc = model.evaluate(val_generator)
print(f"Validation Accuracy: {val_acc:.4f}")
print(f"Validation Loss: {val_loss:.4f}")
```
